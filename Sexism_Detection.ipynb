```python
from spellchecker import SpellChecker

def correct_misspelled_text(text):
    spell = SpellChecker()
    words = text.split()  # Split the text into words
    corrected_words = []

    for word in words:
        corrected_word = spell.correction(word)
        if corrected_word is not None:
            corrected_words.append(corrected_word)

    corrected_text = ' '.join(corrected_words)
    return corrected_text

# Example usage
misspelled_text = "YOU LOSE BIATCHES"
corrected_text = correct_misspelled_text(misspelled_text)
print(misspelled_text)
print(corrected_text)
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
def clean_text(text):
    # Tokenize the text
    tokens = word_tokenize(text.lower())
    # Remove punctuation
    table = str.maketrans('', '', string.punctuation)
    tokens = [w.translate(table) for w in tokens]
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
    # Return the cleaned tokens as a string
    return ' '.join(tokens)
```

2- tweeter is social media that people can use emoji in theire text. so we need to first remmove them and replace with some text that reprensent the emoji it self. so we use emoji library and replace them with theire description.
  😉 --> winking face

```python
import emoji
def emoji2concat_description(text):
    emoji_list = emoji.emoji_list(text)
    ret = emoji.replace_emoji(text, replace='').strip()
    for json in emoji_list:
        this_desc = ' '.join(emoji.EMOJI_DATA[json['emoji']]['en'].split('_')).strip(':')
        ret += ' ' + this_desc
    return ret
```

This means that any less frequent words will be ignored.pad_sequences ensures all sequences in a list have the same length of 20.

9.Creating a simple LSTM Model

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

# Build the neural network model
model = Sequential()
model.add(Embedding(10000, 20))
model.add(LSTM(32))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])

# Train the model
LSTMhistory = model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_val, y_val))
```

```python
from keras.models import Model
from keras.layers import Input, Embedding, LSTM, Dense, Dot, Activation,concatenate
import keras.backend as K

def attention(inputs, TIME_STEPS):
    input_dim = int(inputs.shape[2])
    a = Dense(TIME_STEPS, activation='softmax')(inputs)
    output_attention_mul = Dot(axes=1)([a, inputs])
    return output_attention_mul
TIME_STEPS = 20  # Define the number of time steps in your input sequence
EMBEDDING_DIM = 20
HIDDEN_UNITS = 32
# Define input layer
input_layer = Input(shape=(TIME_STEPS,))
embedding_layer = Embedding(10000, EMBEDDING_DIM)(input_layer)
# Define LSTM layer
lstm_layer = LSTM(HIDDEN_UNITS, return_sequences=True)(embedding_layer)
# Apply attention mechanism
attention_output = attention(lstm_layer, TIME_STEPS)
# Combine attention output with LSTM output
merged = concatenate([lstm_layer, attention_output], axis=-1)
# Apply final Dense layer
output_layer = Dense(1, activation='sigmoid')(merged)
# Build the model
attentionModel = Model(inputs=input_layer, outputs=output_layer)
attentionModel.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
# Train the model
attention_history = attentionModel.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_val, y_val))
```

```python
from keras.models import Sequential, Model
from keras.layers import Embedding, LSTM, Dense, Bidirectional, Dropout, Input, Dot, Activation, Concatenate
import keras.backend as K
from keras.regularizers import l2
from keras.callbacks import EarlyStopping
def attention(inputs):
    a = Dense(TIME_STEPS, activation='softmax')(inputs)
    output_attention_mul = Dot(axes=1)([a, inputs])
    return output_attention_mul
TIME_STEPS = 20
EMBEDDING_DIM = 20
HIDDEN_UNITS = 8
DROPOUT_RATE = 0.5

# Define input layer
input_layer = Input(shape=(TIME_STEPS,))
embedding_layer = Embedding(10000, EMBEDDING_DIM)(input_layer)

# Use Bidirectional LSTM layer
lstm_layer = Bidirectional(LSTM(HIDDEN_UNITS, return_sequences=True))(embedding_layer)
lstm_layer = Dropout(DROPOUT_RATE)(lstm_layer)  # Apply dropout for regularization

# Apply attention mechanism
attention_output = attention(lstm_layer)

# Combine attention output with LSTM output
merged = Concatenate(axis=-1)([lstm_layer, attention_output])

# Apply final Dense layer
output_layer = Dense(1, activation='sigmoid', kernel_regularizer=l2(0.01))(merged)

# Build the model
attentionModel = Model(inputs=input_layer, outputs=output_layer)
attentionModel.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Train the model
attention_history = attentionModel.fit(X_train, y_train, epochs=15, batch_size=128, validation_data=(X_val, y_val), callbacks=[early_stopping])
```

```python
training_loss = attention_history.history['loss']
validation_loss = attention_history.history['val_loss']
training_accuracy = attention_history.history['acc']
validation_accuracy = attention_history.history['val_acc']

# Create epochs range
epochs = range(1, len(training_loss) + 1)

# Plot loss
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs, training_loss, 'b', label='Training Loss')
plt.plot(epochs, validation_loss, 'r', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, training_accuracy, 'b', label='Training Accuracy')
plt.plot(epochs, validation_accuracy, 'r', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertForSequenceClassification, AdamW
from sklearn.model_selection import train_test_split

# Load pre-trained BERT model and tokenizer
modelBert = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
tokenizerBert = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize and encode the text data
texts = Training_Set1['text_clean'].tolist()
labels = Training_Set1['label_sexist'].tolist()

# Split data into train and test sets
texts_train, texts_test, labels_train, labels_test = train_test_split(texts, labels, test_size=0.2, random_state=42)

# Tokenize and encode the train data using the tokenizer
inputs_train = tokenizerBert(texts_train, padding=True, truncation=True, return_tensors='tf', max_length=20)
labels_train = tf.constant(labels_train, dtype=tf.int32)

# Tokenize and encode the test data using the tokenizer
inputs_test = tokenizerBert(texts_test, padding=True, truncation=True, return_tensors='tf', max_length=20)
labels_test = tf.constant(labels_test, dtype=tf.int32)

# Set up optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-5)

# Create a TensorFlow dataset for training
dataset_train = tf.data.Dataset.from_tensor_slices((inputs_train['input_ids'], inputs_train['attention_mask'], labels_train))
dataset_train = dataset_train.batch(16)

# Training loop
for epoch in range(5):
    for batch in dataset_train:
        input_ids, attention_mask, labels = batch
        with tf.GradientTape() as tape:
            logits = modelBert({'input_ids': input_ids, 'attention_mask': attention_mask}, training=True).logits
            loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)(labels, logits)
        gradients = tape.gradient(loss, modelBert.trainable_variables)
        optimizer.apply_gradients(zip(gradients, modelBert.trainable_variables))
```

```python
test_dataset = tf.data.Dataset.from_tensor_slices((inputs_test['input_ids'], inputs_test['attention_mask'], labels_test))
test_dataset = test_dataset.batch(16)

total_correct = 0
total_samples = 0

for batch in test_dataset:
    input_ids, attention_mask, labels = batch
    logits = modelBert({'input_ids': input_ids, 'attention_mask': attention_mask}).logits
    predicted_labels = tf.cast(tf.argmax(logits, axis=1), tf.int32)
    total_correct += tf.reduce_sum(tf.cast(predicted_labels == tf.cast(labels, tf.int32), tf.int32))
    total_samples += labels.shape[0]

accuracy = total_correct / total_samples
print(f'Test Accuracy: {accuracy}')
```

```python
import numpy as np

y_pred_proba = []
y_true = []

for batch in test_dataset:
    input_ids, attention_mask, labels = batch
    logits = modelBert({'input_ids': input_ids, 'attention_mask': attention_mask}).logits
    predicted_probs = tf.nn.softmax(logits, axis=-1)[:, 1]  # Probabilities for the positive class
    y_pred_proba.extend(predicted_probs.numpy())
    y_true.extend(labels.numpy())

y_pred_proba = np.array(y_pred_proba)
y_true = np.array(y_true)
```
